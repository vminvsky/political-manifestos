```python
df=pd.read_csv("sample_manifesto.csv")
```

```python
df["lower_text"]=df["text"].str.lower()
```

```python
import nltk
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
df
```

```python
stop_words=stopwords.words("english")
```

```python
from string import punctuation, digits
import re

def preProcessing(text):
 
  return df_remove_punc(df_remove_digits((text)))

def df_remove_digits(text):
    text=text.strip()
    remove_digits = str.maketrans('', '', digits)
    return text.translate(remove_digits)


def df_remove_punc(text):
    regex = re.compile('[%s]' % re.escape(punctuation))
    return regex.sub('', text)

```

```python
%%time
df["preProcessing"] = df["lower_text"].map(preProcessing)
df.head()
```

```python
df
```

```python
df["preProcessing"]= df["preProcessing"].apply(lambda x: ' '.join([item for item in x.split() if item not in stop_words]))
```

```python
df
```

```python
def word_count(sent):
    word2cnt = dict()
    for word in sent.split():
        if word in word2cnt: word2cnt[word] += 1
        else: word2cnt[word] = 1
    return word2cnt
```

```python
df['word_count'] = df['preProcessing'].apply(word_count)
```
